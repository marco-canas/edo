```python
import numpy as np 
r1 = float(r1)
r2 = float(r2) 
xi1 = np.array(xi1)
xi2 = np.array(xi2) 
```

```python
def x_super1(t):
    return xi1*np.exp(r1*t) 

def x_super2(t):
    return xi2*np.exp(r2*t)  
```
